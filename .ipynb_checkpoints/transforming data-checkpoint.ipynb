```python
import numpy as np
import math

samples = np.load("3_QHO_1mil_samples.npy")
#we break the xList into real and imag parts and then reshape the array so that 
#the last axis picks between those two elements
formatted_samples_3 = np.transpose(np.array([np.real(samples),np.imag(samples)]),[1,2,0])
```

```python
import random

#picking random samples from the total 500k samples
randInd = np.array(random.sample(range(0,len(samples)),20000))
#randInd = np.random.randint(0,len(formatted_samples_3),50000)
formatted_samples_3 = np.array(formatted_samples_3[randInd,:,:])

#applying mirror symmetry
temp = np.zeros((2*len(formatted_samples_3),3,2))
temp[:len(formatted_samples_3),:,0] = formatted_samples_3[:,:,0]
temp[:len(formatted_samples_3),:,1] = formatted_samples_3[:,:,1]
temp[len(formatted_samples_3):,:,0] = formatted_samples_3[:,:,0]
temp[len(formatted_samples_3):,:,1] = -formatted_samples_3[:,:,1]
formatted_samples_3 = np.array(temp)

radial_samples_3 = np.zeros(np.shape(formatted_samples_3))
#radial_samples_3[:,:,0]=np.sqrt(formatted_samples_3[:,:,0]**2 + formatted_samples_3[:,:,1]**2)
#radial_samples_3[:,:,1]=np.mod(np.arctan2(formatted_samples_3[:,:,1],formatted_samples_3[:,:,0]),2*np.pi)
com_samples_3 = np.zeros((len(formatted_samples_3),2))
com_samples_3[:,0]=np.sum(formatted_samples_3[:,:,0],axis=1)/3
com_samples_3[:,1]=np.sum(formatted_samples_3[:,:,1],axis=1)/3

com_rad = np.zeros(len(formatted_samples_3))
com_rad[:] = np.sqrt(com_samples_3[:,0]**2 + com_samples_3[:,1]**2)

weight_com_3 = np.zeros(len(formatted_samples_3))
weight_com_3[:] = np.exp(3*com_rad[:]**2)
#print(formatted_samples_3[temp2][1])
#print(com_rad[temp2][1])


#shifting samples
for i in range(3):
    formatted_samples_3[:,i,0] = formatted_samples_3[:,i,0] - com_samples_3[:,0]
    formatted_samples_3[:,i,1] = formatted_samples_3[:,i,1] - com_samples_3[:,1]


#finding radial version of shifted samples
radial_samples_3[:,:,0]=np.array(np.sqrt(formatted_samples_3[:,:,0]**2 + formatted_samples_3[:,:,1]**2))
radial_samples_3[:,:,1]=np.array(np.mod(np.arctan2(formatted_samples_3[:,:,1],formatted_samples_3[:,:,0]),2*np.pi))

#duplicating sample 3 times and rotating so that each point is rotated to x-axis once
radial_samples_3_1 = np.array(radial_samples_3)
radial_samples_3_1[:,:,1] = np.mod(radial_samples_3_1[:,:,1] - radial_samples_3_1[:,0,1,np.newaxis]*np.ones((len(formatted_samples_3),3)),2*np.pi)

radial_samples_3_2 = np.array(radial_samples_3)
radial_samples_3_2[:,:,1] = np.mod(radial_samples_3_2[:,:,1] - radial_samples_3_2[:,1,1,np.newaxis]*np.ones((len(formatted_samples_3),3)),2*np.pi)

radial_samples_3_3 = np.array(radial_samples_3)
radial_samples_3_3[:,:,1] = np.mod(radial_samples_3_3[:,:,1] - radial_samples_3_3[:,2,1,np.newaxis]*np.ones((len(formatted_samples_3),3)),2*np.pi)

radial_samples_3 = np.concatenate((radial_samples_3_1,radial_samples_3_2,radial_samples_3_3))

#permuting radial samples 
idx=np.argsort(radial_samples_3[:,:,1], axis=1)
radial_samples_3 = np.take_along_axis(radial_samples_3,idx[:,:,None], axis=1)

formatted_samples_3 = np.zeros(np.shape(radial_samples_3))
formatted_samples_3[:,:,0]=radial_samples_3[:,:,0]*np.cos(radial_samples_3[:,:,1])
formatted_samples_3[:,:,1]=radial_samples_3[:,:,0]*np.sin(radial_samples_3[:,:,1])


#del radial_samples_3
del com_samples_3
```

```python
#this is the code for loading all the data and storing it together, just rotating one particle to x-axis

learning_num = 6
qho_samples = np.ndarray([(learning_num-2)*1000000,10], dtype = 'complex64')
lll_samples = np.ndarray([(learning_num-2)*1000000,10], dtype = 'complex64')
                          
for i in range(3,learning_num+1):
    samples = np.load(f"{i}_QHO_1mil_samples.npy")
    trans_samples = np.abs(samples)*np.exp(1j*(np.angle(samples)-np.angle(samples[:,0][:,np.newaxis])))
    qho_samples[int((i-3)*1000000):int((i-3+1)*1000000),:np.shape(samples)[1]] = trans_samples
                              
for i in range(3,learning_num+1):
    samples = np.load(f"{i}_LLL_1mil_samples.npy")
    trans_samples = np.abs(samples)*np.exp(1j*(np.angle(samples)-np.angle(samples[:,0][:,np.newaxis])))
    lll_samples[int((i-3)*1000000):int((i-3+1)*1000000),:np.shape(samples)[1]] = trans_samples
    
combined_samples = np.concatenate([qho_samples, lll_samples], axis=0)

qho_labels = np.zeros((qho_samples.shape[0], 1), dtype=int)  # 0s for qho_samples
lll_labels = np.ones((lll_samples.shape[0], 1), dtype=int)   # 1s for lll_samples
combined_labels = np.concatenate([qho_labels, lll_labels], axis=0)

#shuffling indices to randomize qho/lll order
indices = np.arange(combined_samples.shape[0])
np.random.shuffle(indices)

#shuffling samples/labels
combined_samples = combined_samples[indices]
combined_labels = combined_labels[indices]
```

```python
import numpy as np
import math
import h5py
import torch
from torch.utils.data import Dataset
from torch.utils.data import DataLoader
from torch.utils.data import TensorDataset
from torch.utils.data.dataset import random_split

coord = 10

#this is the code for loading all the data, shifting CoM, creating n rotated samples per original sample, and storing it together in h5 files 
#used to create "QHO_trans_samples.h5" and "LLL_trans_samples.h5"
learning_num = 6

file_name = "QHO_trans_samples.h5"
with h5py.File(file_name, 'a') as f:
    rows = sum([i for i in range(3,learning_num+1)])*1000000
    cols = 10
    dset = f.create_dataset('qho_samples', (rows, cols), dtype = 'complex64')
    for k in range(3,learning_num+1):
        samples = np.load(f"{k}_QHO_1mil_samples.npy")
        com_samples = np.sum(samples,axis = 1)/k
        samples = samples - com_samples[:,np.newaxis]
        trans_samples = np.ndarray([np.shape(samples)[0],np.shape(samples)[1]], dtype = 'complex64')  
        ind_shift = sum([i for i in range(3,k)])*1000000
        for i in range(np.shape(samples)[1]):
            trans_samples[:] = np.abs(samples)*np.exp(1j*(np.angle(samples)-np.angle(samples[:,i][:,np.newaxis])))
            trans_samples = np.roll(trans_samples,-i,1)
            dset[ind_shift+i*1000000:ind_shift+(i+1)*1000000,:np.shape(trans_samples)[1]] = trans_samples
        
                
file_name = "LLL_trans_samples.h5"
with h5py.File(file_name, 'a') as f:
    rows = sum([i for i in range(3,learning_num+1)])*1000000
    cols = 10
    dset = f.create_dataset('lll_samples', (rows, cols), dtype = 'complex64')
    for k in range(3,learning_num+1):
        samples = np.load(f"{k}_LLL_1mil_samples.npy")
        com_samples = np.sum(samples,axis = 1)/k
        samples = samples - com_samples[:,np.newaxis]
        trans_samples = np.ndarray([np.shape(samples)[0],np.shape(samples)[1]], dtype = 'complex64')  
        ind_shift = sum([i for i in range(3,k)])*1000000
        for i in range(np.shape(samples)[1]):
            trans_samples[:] = np.abs(samples)*np.exp(1j*(np.angle(samples)-np.angle(samples[:,i][:,np.newaxis])))
            trans_samples = np.roll(trans_samples,-i,1)
            dset[ind_shift+i*1000000:ind_shift+(i+1)*1000000,:np.shape(trans_samples)[1]] = trans_samples

        
```
